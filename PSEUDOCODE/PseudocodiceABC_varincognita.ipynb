and we choose as model:

\begin{equation*}
\begin{aligned}
    y_i | \mu &\sim ^{iid} \mathcal{N}(\mu, \sigma ^2)  \\
    \mu| \tau & \sim \mathcal{N}(\mu_0, 1/k\tau) \\
    \tau & \sim \mathcal{Ga}(a,b) \\
    \pi( \mu,\tau) =\pi( \mu|\tau)*\pi(\tau)
\end{aligned}
\end{equation*}




$\sigma_{obs}$ known

$\mu_0, \sigma_0, a ,b$ fixed

PROSEGUIMENTO ALGORITMO:

We choose a vector of sufficient summary statistics: $ S(y)$

We choose a gaussian Kernel function $Kh(u)$

MCMC ABC:

- Compute $s_{obs} = S(y_{obs})$

- Generate $\mu^{(0)} ∼  \pi(\mu)$ and $\tau^{(0)} ∼  \pi(\tau)$ from prior density.
- Generate two samples of 1000 observations such that $y ∼ \mathcal{N}(\mu^{(0)}, \tau^{(0)})$ 

- Compute $s^{(0)}=S(y)$

Untill $Kh(||s^{(0)}−s_{obs}||)>0$
- Generate $\mu^{(0)} ∼  \pi(\mu)$ and $\tau^{(0)} ∼  \pi(\tau)$ from prior density.
- Generate a sample of 1000 observations such that $y_{i} ∼ \mathcal{N}(\mu^{(0)}, \tau^{(0)})$

- Compute $s^{(0)}=S(y)$








For i = 1,...,N:
1. Generate $\mu^{(i)}$ from $\mathcal{N}(\mu^{(i-1)}, 1)$ and $\tau^{(i)}$ from $\mathcal{N}(\tau^{(i)},1)$
2. Generate a sample of 1000 observations $y$ ∼ $p(y|(\mu^{(i)},\tau^{(i)})$ 
3. Compute $s^{(i)}=S(y)$ 
4. Accept $\mu^{(i)},\tau^{(i)}$ with probability $Kh(||s^{(i)}−s_{obs}||)π(\mu^{(i)},\tau^{(i)})/Kh(||s^{(i-1)}−s_{obs}||)\pi(\mu^{(i-1)},\tau^{(i-1)})$

$\theta=(\mu, \tau)$
