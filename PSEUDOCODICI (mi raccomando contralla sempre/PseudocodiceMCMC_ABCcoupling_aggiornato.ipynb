We generate 1000 observations from a Gaussian distribution:

\begin{equation*}
\begin{aligned}
    y_{obs} &\sim \mathcal{N}(\mu_{obs}, σ_{obs} ^2) \\
\end{aligned}
\end{equation*}

and we choose as model:

\begin{equation*}
\begin{aligned}
    y_i | \mu &\sim ^{iid} \mathcal{N}(\mu, \sigma_{obs} ^2)  \\
    \mu & \sim \mathcal{N}(\mu_0, \sigma_0^2)
\end{aligned}
\end{equation*}

$\sigma_{obs}$ known

$\mu_0, \sigma_0$ fixed

We choose a vector of sufficient summary statistics: $ S(y)$

We choose a gaussian Kernel function $Kh(u)$

MCMC ABC + COUPLING:

- Compute $s_{obs} = S(y_{obs})$

- Generate $θ_{x}^{(0)} ∼  \pi(\mu)$ and $θ_{y}^{(0)} ∼  \pi(\mu)$ from prior density.
- Generate two samples of 1000 observations such that $y_{1i} ∼ \mathcal{N}(θ_{x}^{(0)}, σ_{obs} ^2)$ and  $y_{2j} ∼ \mathcal{N}(θ_{y}^{(0)}, σ_{obs} ^2)$

- Compute $s_{x}^{(0)}=S(y_{1})$ and $s_{y}^{(0)}=S(y_{2})$


Untill $Kh(||s_{x}^{(0)}−s_{obs}||)>0$
- Generate $θ_{x}^{(0)} ∼  \pi(\mu)$ from prior density.
- Generate a sample of 1000 observations such that $y_{1i} ∼ \mathcal{N}(θ_{x}^{(0)}, σ_{obs} ^2)$

- Compute $s_{x}^{(0)}=S(y_{1})$


Untill $Kh(||s_{y}^{(0)}−s_{obs}||)>0$
- Generate $θ_{y}^{(0)} ∼  \pi(\mu)$ from prior density.
- Generate a sample of 1000 observations such that $y_{2j} ∼ \mathcal{N}(θ_{y}^{(0)}, σ_{obs} ^2)$ 
- Compute $s_{y}^{(0)}=S(y_{2})$







For i = 1,...,N:
1. Generate [$θ_{x}^{(i)},θ_{y}^{(i)}$] from a maximal coupling given [$θ_{x}^{(i-1)},θ_{y}^{(i-1)}$].
2. Generate a sample of 1000 observations $y_{1}$ ∼ $p(y|θ_{x}^{(i)})$ and a sample of 1000 observations $y_{2}$ ∼ $p(y|θ_{y}^{(i)})$.
3. Compute $s_{x}^{(i)}=S(y_{1})$ and $s_{y}^{(i)}=S(y_{2})$
4. Accept $θ_{x}^{(i)}$ with probability $Kh(||s_{x}^{(i)}−s_{obs}||)π(θ_{x}^{(i)})/Kh(||s_{x}^{(i-1)}−s_{obs}||)π(θ_{x}^{(i-1)})$ and accept $θ_{y}^{(i)}$ with probability $Kh(||s_{y}^{(i)}−s_{obs}||)π(θ_{y}^{(i)})/Kh(||s_{y}^{(i-1)}−s_{obs}||)π(θ_{y}^{(i-1)})$ 



IN THIS CASE WE OBTAIN TWO CHAINS WHICH DON'T STAY TOGETHER AFTER MEETING POINT


Output:

Two sets of parameter vectors: $θ_{x}(1),...,θ_{x}^{(N)}∼ π_{ABC} (θ|y_{obs})$ and $θ_{y}^{(1)},...,θ_{y}^{(N)} ∼ π_{ABC} (θ|y_{obs})$.
